```python
import matplotlib.pyplot as plt
```

```python
n = [100000, 200000, 300000, 400000, 500000, 600000, 700000, 800000, 900000, 1000000]
```

```python
isPrime0 = [242, 940, 2046, 2870, 4350, 6142, 8264, 10668, 13406, 16398]

plt.plot(n, isPrime0, 'b-', label='Line Plot')
plt.scatter(n, isPrime0, color='black', label='Data Points')
plt.xlabel('n')
plt.ylabel('Time')
plt.title('n vs isPrime0')
plt.grid(True)
plt.ticklabel_format(style='plain', axis='y')
plt.legend()
plt.show()
```

```python
isPrime1 = [7, 10, 13, 22, 29, 38, 47, 55, 64, 76]
isPrime2 = [4, 7, 8, 8, 10, 14, 18, 21, 24, 28]

plt.plot(n, isPrime1, 'b-', label='isPrime1')
plt.scatter(n, isPrime1, color='black', label='Data Points isPrime1')
plt.plot(n, isPrime2, 'b-', color='orange', label='isPrime2')
plt.scatter(n, isPrime2, color='orange', label='Data Points isPrime2')
plt.xlabel('n')
plt.ylabel('Time')
plt.title('n vs isPrime1 and isPrime2')
plt.grid(True)
plt.ticklabel_format(style='plain', axis='y')
plt.legend()
plt.show()
```
