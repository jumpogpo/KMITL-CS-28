```python
import random, math
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
r = 2.0
N = 5000
d = { "Trials": [], "Pi": [] }
```

```python
for i in range(1, N):
  circle_p = 0
  for x in range(i):
    x = random.uniform(-2, r)
    y = random.uniform(-2, r)

    x2 = x ** 2
    y2 = y ** 2

    if math.sqrt(x2 + y2) <= r:
      circle_p += 1
  d["Trials"].append(i)
  d["Pi"].append((circle_p / i) * 4)
```

```python
df = pd.DataFrame(data = d)
```

```python
df
```

```python
plt.figure(figsize = (10,7))
plot = sns.scatterplot(x="Trials", y="Pi", s=30, marker="o", data=df)
plot.set(title='Monte Carlo Simulation to Estimate Value of Pi', xlabel="Number of Trials", ylabel="Value of Pi")
plt.axhline(y=3.14, color='r', linestyle='-')
plt.show()
```

```python
sns.set_style('darkgrid')
fig = plt.figure(figsize = (10,6))
sns.histplot(df, x="Pi",color='orange');
```
