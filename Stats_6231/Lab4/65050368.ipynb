```python
import numpy as np
import pandas as pd
import scipy.stats as stats
df = pd.read_csv('BloodPressure.csv')
```

```python
dfMale = df.loc[df["sex"] =="Male"]
```

```python
dfMale
```

```python
dfFemale = df.loc[df["sex"] =="Female"]
```

```python
dfFemale
```

```python
dfMale[['bp_before','bp_after']].describe()
```

```python
dfFemale[['bp_before','bp_after']].describe()
```

```python
significance = 0.05
stat,p_value = stats.ttest_rel(dfMale['bp_before'], dfMale['bp_after'])
```

```python
print('t_stat: %0.4f, p_value: %0.4f' % (stat, p_value))
if p_value < significance:
   print ("Reject the null hypothesis")
else:
   print ("Accpet the null hypothesis") 
```

```python
significance = 0.05
stat,p_value = stats.ttest_rel(dfFemale['bp_before'], dfFemale['bp_after'])
```

```python
print('t_stat: %0.4f, p_value: %0.4f' % (stat, p_value))
if p_value < significance:
   print ("Reject the null hypothesis")
else:
   print ("Accpet the null hypothesis") 
```
